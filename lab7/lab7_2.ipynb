```python
## Adjust/Remove the Outlier present in the Profit coloumn by using IQR
## IQR = Q3-Q1
```

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
lower_thres = Q1 - (1.5*IQR)
print('The value of lower threshld is :',round(lower_thres,2))
```

```python
df[df['Profit']<lower_thres]
```

```python
## Will Change the row in order to avoid outlier
df['Profit'] = np.where(df['Profit']<lower_thres,lower_thres,df['Profit'])
```

```python
## Box plot of Profit data---------after avoiding outlier to check
plt.figure(figsize=(6, 2))
plt.boxplot(df['Profit'], vert=False, patch_artist=True)
plt.title('Box Plot of Profit')
plt.xlabel('Profit')
plt.show()
```

## Model Fitting

### Method----1(a)
#### Using Simple Linear Regression (R&D spend and profit)

```python
model_data_1a = df[['R&D Spend','Profit']]
print('The model data is given as:\n', model_data_1a.head())
```

```python
# Correlation b/w R&D_Spending and Profit
correlation = df[['R&D Spend', 'Profit']].corr()
correlation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_1a = model_data_1a['R&D Spend']
y_1a = model_data_1a['Profit']
```

```python
# Sampling the data to train/test data
x_train_1a,x_test_1a,y_train_1a,y_test_1a = train_test_split(x_1a, y_1a, test_size=0.2, random_state=42)
```

```python
## The shape of data after distributing it into train and test category
x_train_1a.shape,y_train_1a.shape
```

```python
x_test_1a.shape,y_test_1a.shape
```

```python
# Reshaping of the data as it is a simple Linear Regression
x_train_1a = x_train_1a.values.reshape(-1, 1)
y_train_1a = y_train_1a.values.reshape(-1, 1)
```

```python
## Model Building using `sklearn`
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
print(model)
```

```python
model_fit_1a = model.fit(x_train_1a, y_train_1a)
model_fit_1a
```

```python
# Print Coefficient
print(f"The coefficient of the fitted data is: {model.coef_[0]}")
```

```python
# Print Intercept of the fitting
print(f"The Intercept of the fitted data is: {model.intercept_[0]}")
```

```python
## Model Prediction
x_test_1a = x_test_1a.values.reshape(-1,1)
```

```python
pred_1a = model.predict(x_test_1a)
pred_1a
```

```python
df_pred_1a = pd.DataFrame(pred_1a,columns=['pred'])
df_pred_1a
```

```python
df_pred_1a['Actual'] = y_test_1a.values
df_pred_1a
```

```python
# Error calculation
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from math import sqrt
```

```python
r2_value = r2_score(y_test_1a, pred_1a)

# Display the R2 value
print(f"R-squared (R2) value: {r2_value}")
```

```python
# R square value
r2_value = r2_score(y_test_1a, pred_1a)
print(f"The strength of the model is : {r2_value}")
```

### Method---2
#### Using multiple linear regression

```python
df['State'].unique()
```

```python
# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns = ['State']).astype(int)

df_encoded.head()
```

```python
# Concatenate the two datas
df1 = pd.concat([df, df_encoded], axis=1)
df1.head()
```

```python
# Get the final data
x2 = df1[['R&D Spend','Administration','Marketing Spend','State_California', 'State_Florida', 'State_New York']]
x2.head()
```

```python
y2 = df['Profit']
y2.head()
```

```python
## Sampling of data to train/test data
x_train2, x_test2, y_train2, y_test2 = train_test_split(x2, y2, test_size=0.2, random_state=42)
```

```python
# No need to reshape the data in case of multiple linear regression
```

```python
# Model building
model2 = LinearRegression()
model2.fit(x_train2, y_train2)
```

```python
# Coefficient
print(f"The coefficient of the fitted data is: {model2.coef_}")
```

```python
# Intercept
print(f"The intercept of the fitted data is: {model2.intercept_}")
```

```python
# Model prediction
y_pred2 = model2.predict(x_test2)
y_pred2
```

```python
dd = pd.DataFrame({'predicted': y_pred2.flatten()})
```

```python
y_test2
```

```python
dd['Actual'] = y_test2.values
```

```python
dd
```

```python
# R square value
r_squared = r2_score(y_test2, y_pred2)
r_squared
```

```python
# RMSE
mse = mean_squared_error(y_test2, y_pred2)
rmse = np.sqrt(mse)
rmse
```
